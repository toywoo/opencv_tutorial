```python
import numpy as np, cv2
import matplotlib.pyplot as plt
```

```python
m1 = np.full((3, 6), 10, np.uint8)
m2 = np.full((3, 6), 50, np.uint8)
m_mask = np.zeros(m1.shape, np.uint8)
m_mask[ :, 3:] = 1

m_add1 = cv2.add(m1, m2)
m_add2 = cv2.add(m1, m2, mask=m_mask)

m_div1 = cv2.divide(m1, m2)
m1 = m1.astype(np.float32) # cvt type1
m2 = np.float32(m2) # cvt type2
m_div2 = cv2.divide(m1, m2)

titles = ['m1', 'm2', 'm_mask', 'm_add1', 'm_add2', 'm_div1', 'm_div2']
for title in titles:
    print("%s = \n%s \n" % (title, eval(title)))
```

```python
v1 = np.array([1, 2, 3], np.float32)
v2 = np.array([[1], [2], [3]], np.float32)
v3 = np.array([[1, 2, 3]], np.float32)

v_exp = cv2.exp(v1)
v_log = cv2.log(v1)
m_sqrt = cv2.sqrt(v1)
m_pow = cv2.pow(v1, 3)

# 열벡터에서 행벡터로 바꾸는 방법
print("[exp] = ", v_exp.T)
print("[log] = ", v_log.T)
print("[log] = ", np.ravel(v_log))
print("[sqt] = ", np.ravel(m_sqrt))
print("[pow] = ", m_pow.flatten())
```

```python
x = np.array([1, 2, 3, 5, 10], np.float32)
y = np.array([2, 5, 7, 2, 9]).astype("float32")

mag = cv2.magnitude(x, y)
ang = cv2.phase(x, y)
p_mag, p_ang = cv2.cartToPolar(x, y)

titles = ['x', 'y', 'mag', 'ang', 'p_mag', 'p_ang']
for title in titles:
    print("%s = \n%s \n" % (title, eval(title).flatten()))
```

```python
img1 = np.zeros((300, 300), np.uint8)
img2 = img1.copy()

h, w = img1.shape[:2]
cx, cy = w//2, h//2
cv2.circle(img1, (cx, cy), 100, 255, -1)
cv2.rectangle(img2, (0, 0, cx, h), 255, -1)

img_or = cv2.bitwise_or(img1, img2)
img_and = cv2.bitwise_and(img1, img2)
img_xor = cv2.bitwise_xor(img1, img2)
img_not = cv2.bitwise_not(img1, img2)

titles =  ['img1', 'img2', 'img_or', 'img_and', 'img_xor', 'img_not']

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(8, 6))

for ax, title in zip(axs.flat, titles):
    ax.imshow(eval(title))
    ax.axis("off")
    ax.set_title(title)
```

```python
img1 = cv2.imread("../images/timetable.jpg", cv2.IMREAD_GRAYSCALE)

img2 = np.zeros_like(img1) #서로 크기가 같아야 함

h, w = img2.shape[:2]
cx, cy = w//2, h//2
cv2.circle(img2, (cx, cy), 400, 255, -1)

dif_img1 = cv2.subtract(img1, img2)
dif_img2 = cv2.subtract(np.int16(img1), np.int16(img2))
abs_dif1 = np.absolute(dif_img2).astype('uint8')
abs_dif2 = cv2.absdiff(img1, img2)

titles =  ['img1', 'img2', 'dif_img1', 'dif_img2', 'abs_dif1', 'abs_dif2']

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(8, 6))

for ax, title in zip(axs.flat, titles):
    print(title, eval(title))
    ax.imshow(eval(title))
    ax.axis("off")
    ax.set_title(title)
```

```python
img1 = cv2.imread("../images/timetable.jpg", cv2.IMREAD_COLOR)

img2 = np.zeros_like(img1) #서로 크기가 같아야 함

mean = cv2.mean(img1)
img2 = cv2.add(img1, mean)

img3 = img1.copy()
i  = 0
while i < 200:
    img3[100 + i] = cv2.sort(img1[100 + i], cv2.SORT_EVERY_COLUMN)
    i += 1

dif_img1 = cv2.subtract(img1, img2)
dif_img2 = cv2.subtract(np.int16(img1), np.int16(img2))
abs_dif1 = np.absolute(dif_img2).astype('uint8')
abs_dif2 = cv2.absdiff(img1, img2)

titles =  ['img1', 'img2', 'img3']

fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))

for ax, title in zip(axs.flat, titles):
    ax.imshow(eval(title))
    ax.axis("off")
    ax.set_title(title)
```
