```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

기본 배열 처리 함수

```python
img = cv2.imread("../images/timetable.jpg", cv2.IMREAD_COLOR)
if img is None: raise Exception("영상 파일 읽기 오류 발생")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

x_axis = cv2.flip(img, 0)
y_axis = cv2.flip(img, 1)
xy_axis = cv2.flip(img, -1)
rep_img = cv2.repeat(img, 1, 2)
trans_img = cv2.transpose(img)

titles =  ['img', 'x_axis', 'y_axis', 'xy_axis', 'rep_img', 'trans_img']

imgs = [img, x_axis, y_axis, xy_axis, rep_img, trans_img]
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(8, 6))

for ax, img, title in zip(axs.flat, imgs, titles):
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(title)

plt.tight_layout()
plt.show()    
```

bgr 각각의 원소들을 선언하고 합쳐서 화면에 각 화소에 어떤 bgr 값이 있는지 알게 해줌, 그리고 그걸을 다시 bgr로 나누어 각 화소에 b에 뭐 있는지 ....있는지 알게 해준다.

```python
ch0 = np.zeros((2, 4), np.uint8) + 10
ch1 = np.ones((2, 4), np.uint8) * 20
ch2 = np.full((2,4), 30, np.uint8)

list_bgr = [ch0, ch1, ch2]
merge_bgr = cv2.merge(list_bgr)
split_bgr = cv2.split(merge_bgr)

print("list_bgr", list_bgr, end="\n\n")
print("merge_bgr", merge_bgr, merge_bgr.shape, end="\n\n")
print("split_bgr", split_bgr, np.array(split_bgr).shape, end="\n\n")
```

```python
img = cv2.imread("../images/timetable.jpg", cv2.IMREAD_COLOR)
if img is None: raise Exception("영상 파일 읽기 오류 발생")

splited_img = cv2.split(img)

titles =  ['img', 'blue', 'green', 'red']
imgs = [img, splited_img[0], splited_img[1], splited_img[2]]

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

for ax, img, title in zip(axs.flat, imgs, titles):
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(title)
```
